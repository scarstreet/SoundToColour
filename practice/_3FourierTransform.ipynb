```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import statistics as s

song_path = 'pps.mp3'
song, sr = librosa.load(song_path)
song = song[5000:40000]
```

```python
plt.figure(figsize=(18,8))
plt.title("Waveform of one beat in Time Domain")
librosa.display.waveshow(song, sr=sr, alpha=0.5)
plt.show()
```

```python
ft = sp.fft.fft(song)
magnitude = np.absolute(ft)
freq = np.linspace(0,sr,len(magnitude))
```

```python
plt.figure()
plt.plot(freq[:5000], magnitude[:5000])
plt.ylim((0,500))
plt.xlabel("frequency(Hz)")
plt.ylabel("magnitude")
plt.title("The Frequency Domain of \n One Beat (Max Magnitude at 314Hz)\n")
```
